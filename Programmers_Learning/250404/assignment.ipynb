```python
#2강 실습: (1) 리스트에 새로운 요소 삽입하기
def solution(L, x):
    if L[:1][0] >= x: #모든 원소보다 작거나
        L.insert(0,x)
    elif L[-1:][0] <= x: #모든 원소보다 크거나
        L.append(x)
    else: #그 사이에 있을때
        for i in range(len(L)):
            if L[i] >= x:
                L.insert(i,x)
                break        
    return L

print(solution([20, 37, 58, 72, 91], 65))
print(solution([20, 37, 67, 72, 91], 65))
print(solution([20, 37, 67, 72, 91], 1))
print(solution([20, 37, 67, 72, 91], 99))
```

```python
#2강 실습: (2) 리스트에서 원소 찾아내기
#재귀로 검색
def solution(L, *x):
    if len(x) == 1: # 1개면 처음 호출 2개 이사이면 재귀호출
        answer=[]
        try:
            answer.append(L.index(x[0]))
        except ValueError:
            return [-1] # 처음부터 없을때
        
        #첫자리 = 처음 발견한 값, 이후 값 = 첫자리로부터 몃번째 인덱스 인지
        temp = solution(L[answer[0]+1:], x[0],answer)
        for i in range(1,len(temp)):
            temp[i]=temp[i-1]+temp[i]+1

        return temp
    else:#x[0]:찾을 값 / x[1]:현재까지 찾은 index 리스트
        try:
            x[1].append(L.index(x[0]))
        except ValueError:
            return x[1] # 누적 X 반환
        
        return solution(L[x[1][-1:][0]+1:], x[0],x[1])
    
print(solution([6, 2, 8, 7, 3, 3,3],3))
print(solution([6, 2, 8, 7, 3],3))
print(solution([6, 2, 8, 7, 4],3))
```

```python
#3강 실습: 이진 탐색 구현해보기
def solution(L, x):
    
    
    def loop_function(L,x,s,e):                 #재귀처리로 바꾸기 위해 별도 내부함수로 처리
        if s>e:                                 #재귀후 모든 값을 비교했음에도 안나왔을때 -1 리턴
            return -1 
        
        mid = (s+e)//2                          #중간 index 구하기
        #print(mid,s,e,sep="/")
        
        if L[mid] == x:                         #구한 중간Index를 이용해 찾고자 하는 값인지 확인
            return mid
        elif L[mid] > x:
            return loop_function(L, x,s,mid-1)  #중간 Index의값이 찾고자 하는 값보다 작다면 시작점~중간점-1까지 재 조회(재귀)
        elif L[mid] < x:
            return loop_function(L, x,mid+1,e)  #중간 Index의값이 찾고자 하는 값보다 크다면 중간점+1~종료점까지 재 조회(재귀)
    
    return loop_function(L,x,0,len(L)-1)

L = [2, 3, 5, 6, 9, 11, 15, 17, 19]
x = 3
print(solution(L,x))
L = [2, 5, 7, 9, 11]
x = 4
print(solution(L,x))
L = [1, 2, 3]
x = 3
print(solution(L,x))
```

```python
#8강: 연결 리스트(Linked List) (2) linkedlist.py
class Node:

    def __init__(self, item):
        self.data = item
        self.next = None


class LinkedList:

    def __init__(self):
        self.nodeCount = 0
        self.head = None
        self.tail = None


    def __repr__(self):
        if self.nodeCount == 0:
            return 'LinkedList: empty'

        s = ''
        curr = self.head
        while curr is not None:
            s += repr(curr.data)
            if curr.next is not None:
                s += ' -> '
            curr = curr.next
        return s


    def getAt(self, pos):
        if pos < 1 or pos > self.nodeCount:
            return None

        i = 1
        curr = self.head
        while i < pos:
            curr = curr.next
            i += 1

        return curr


    def insertAt(self, pos, newNode):
        if pos < 1 or pos > self.nodeCount + 1:
            return False

        if pos == 1:
            newNode.next = self.head
            self.head = newNode

        else:
            if pos == self.nodeCount + 1:
                prev = self.tail
            else:
                prev = self.getAt(pos - 1)
            newNode.next = prev.next
            prev.next = newNode

        if pos == self.nodeCount + 1:
            self.tail = newNode

        self.nodeCount += 1
        return True


    def getLength(self):
        return self.nodeCount


    def traverse(self):
        result = []
        curr = self.head
        while curr is not None:
            result.append(curr.data)
            curr = curr.next
        return result


    def concat(self, L):
        self.tail.next = L.head
        if L.tail:
            self.tail = L.tail
        self.nodeCount += L.nodeCount



a= Node(1)
b= Node(2)
c= Node(3)
d= Node(4)
e= Node(5)

L=LinkedList()

L.insertAt(1,a)
L.insertAt(1,b)
L.insertAt(1,c)
L.insertAt(1,d)
L.insertAt(1,e)

print(L)
```

```python
#12강 실습: 중위표현 수식 -> 후위표현 수식
class ArrayStack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]

prec = {
    '*': 3, '/': 3,
    '+': 2, '-': 2,
    '(': 1
}

def solution(S):
    opStack = ArrayStack()
    answer =""
    for c in S:
        if c in "*/+-()": #연산자
            if c == "(":
                opStack.push(c)
            elif c == ")":
                while not opStack.peek() == "(":
                    answer+=opStack.pop()
                opStack.pop()#"(" 버림
            else:
                while (not opStack.isEmpty() and opStack.peek() != "(" and prec[opStack.peek()] >= prec[c]):
                    answer += opStack.pop()
                opStack.push(c)
        else:#피연산자
            answer+=c
    while not opStack.isEmpty():
        answer+=opStack.pop()

    return answer


print(1,solution("(A+B)*(C+D)"),"AB+CD+*")
print(2,solution("A*B+C"),"AB*C+")
print(3,solution("A+B*C"),"ABC*+")
print(4,solution("A+B+C"),"AB+C+")
print(5,solution("(A+B)*C"),"AB+C*")
print(6,solution("A*(B+C)"),"ABC+*")
```
